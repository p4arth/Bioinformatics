```python
DNA_ReverseCompliment ={'A': 'T','T':'A', 'G':'C', 'C':'G'}
```

```python
def reverse_compliment(seq):
  return ''.join([DNA_ReverseCompliment[nuc] for nuc in seq])[::-1]
```
